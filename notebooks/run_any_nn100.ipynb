```python
def load_preselected_queries():
    cats_and_queries = np.loadtxt("../selected_queries.csv", delimiter=",", dtype=int)
    return cats_and_queries[:,0], cats_and_queries[:,1]
```

```python
from sisap2023.utils.distances import get_euc_dists
from sisap2023.utils.count_cats import *
from sisap2023.metrics.msed_class import MSED
import numpy as np

nth_query = 6

cats, queries = load_preselected_queries()

query = queries[nth_query] 

dists = get_euc_dists(query, data)
closest_indices = np.argsort(dists)  # the closest images to the query
    
best_k_for_one_query = closest_indices[0:100]  # the k closest indices in data to the query
category = get_topcat(query, sm_data)
best_k_categorical = get_best_cats_in_subset(category, best_k_for_one_query, sm_data)  # the closest indices in category order - most peacocky peacocks etc.
poly_query_indexes = best_k_categorical[0:5]  # These are the indices that might be chosen by a human
poly_query_data = data[poly_query_indexes]  # the actual datapoints for the queries

base = MSED(np.array(poly_query_data))
complexities = base.query(data)

print(categories[category])

```

```python

complexities = complexities.flatten()   # <<< these are all matrices of matrices shouldn't be.
indices = np.argsort(complexities)      # lowest is first
top_100_poly = indices[0:100]
```
